```python
#30dayspythonchallenge_Day13
# is a compact way of creating a list from a sequence
# One way
lang = "python"
print(list(lang))
print((type(lang)))

# list comprehension 
lang = "money"   # changing a string to a list using comprehension
lst = [i for i in lang]
print(type(lst))
print(lst)

#to generate list of numbers
num = [i for i in range(11)]
print(num)

#
num = [ i * i for i in range(4)]
print(num)

# for tuple
num = [(i, i* i ) for i in range(4)]
print(num)

# list comprehension can be combined with if expression
num = [i for i in range(11) if i % 2 == 0]
print(num)
num = [ i for i in range(11) if i % 2 != 0]
print(num)

# just by the way
def num():
    for i in range(11):
        if i % 2 == 0:
           print(i) 
    
print(num())

# 
numbers = [-8, -7, -3, -1, 0, 1, 3, 4, 5, 7, 6, 8, 10]
num = [i for i in numbers if i % 2 == 0 and i > 0]
print(num)

numbers = [-8, -7, -3, -1, 0, 1, 3, 4, 5, 7, 6, 8, 10]
num = [i for i in numbers if i % 2 != 0 and i > 0]
print(num)


#Lambda function is use for creating anonymous function inside another function
   #syntax
   #  x = lambda a,b,c: a+b+c
   #  print(x(a,b,c))
   
#1
def add(a,b):
    return a +b
print(add(2,3))

#using lambda function
x = lambda a,b: a + b
print(x(2,5))

#self invoking lamda
square = lambda x: x**2
print(square(2))
num = lambda a, b, c: a ** 2 - 3 * b + 4 * c
print(num(5,5,3))

# lambda function inside another function
def power(x):
    return lambda n: x ** n
cube = power(2)(3)
print(cube)
total = power(2)(5)
print(total)
```
