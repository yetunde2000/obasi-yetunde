```python
## Method to Modify Class Default Values
class Person:
    def __init__(self, first ="Yetunde", last = "Obasi", Age = 25, country = "south korea", city = "daegu"):
        self.first = first
        self.last = last
        self.Age = Age
        self.country = country
        self.city = city
        self.skills = []
    def person_info(self):
        return f'{self.first} {self.last} is {self.Age} years old. She lives in {self.country} {self.city}'
    def add_skill(self, skill):
        self.skills.append(skill)
p = Person()
print(p.person_info())
p.add_skill('HTML')
p.add_skill('Java')
p.add_skill('Python')
print(p.skills)

# Inheritance
# Using inheritance we can reuse parent class code. Inheritance allows us to define a class that inherits all the methods and properties from parent class. 
class Student(Person):
    pass
s1 = Student('Eyob', 'Yetayeh', 30, 'Finland', 'Helsinki')
s2 = Student('Lidiya', 'Teklemariam', 28, 'Finland', 'Espoo')
print(s1.person_info())
s1.add_skill('LSTM')
s1.add_skill('LLM')
print(s1.skills)
print(s2.person_info())
```

```python
### Exercises: Day 21
### Exercises: Level 1
import statistics 
from collections import Counter
def freq_dist(ages):
    return dict(Counter(ages))
class Statistics:
    def __init__(self, ages):
        self.ages = ages
        self.mean = statistics.mean(ages)
        self.median = statistics.median(ages)
        self.percentile = statistics.quantiles(ages)

ages = [31, 26, 34, 37, 27, 26, 32, 32, 26, 27, 27, 24, 32, 33, 27, 25, 26, 38, 37, 31, 34, 24, 33, 29, 26]
data = Statistics(ages)
print('Count:', len(data.ages))
print('Sum:', sum(data.ages))
print('Min:', min(data.ages))
print('Max:', max(data.ages))
print('Range:',max(data.ages)-min(data.ages))
print('Mean:', statistics.mean(data.ages))
print('Median:', statistics.median(data.ages))
print('Mode:', statistics.mode(data.ages))
print('Variance:', statistics.variance(data.ages))
print('Frequency:', freq_dist(data.ages))


## Exercises: Level 2
class PersonAccount:
    def __init__(self, firstname, lastname, incomes): #, account_info, add_income, add_expense, account_balance):
        self.firstname = firstname
        self.lastname = lastname
        self.incomes = incomes
    
    def per_info(self):
        return f'{self.firstname} {self.lastname} {self.incomes}'
p = PersonAccount('Yetunde', 'Obasi', 100000)
print(p.per_info())

        
```
