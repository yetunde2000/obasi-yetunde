```python
#30dayspythonchallenge_Day5 
# list A list is collection of different data types which is ordered and modifiable(mutable). A list can be empty or it may have different data type items.
lst = list()
print(lst)
fruits = ["banana", "orange", "mango", "lemon"]
vegetables = ["Tomato", "Potato", "Cabbage", "Onion"]
countries = ["Uk", "south korea", "Nigeria", "Sweden"]
print("Fruits:", fruits)
print("Vegetables:", vegetables)
print("Number of countries:", len(countries))

#accessing in list
Lost = ["love", "orange", "mango"]
print(Lost[2])
print(len(Lost) - 1)
```

```python
#accessing list with negative index
fruits = ["orange", "banana", "mango", "lemon"]
print(fruits[-1])
print(fruits[-3])

## Unpacking list items
list = ["kim","lim", "lee", "jang", "Na"]
first_term, second_term, third_term, *rest = list
print(first_term)
print(second_term)
print(third_term)
print(*rest)

#
list = ["cane", "abel", "kimon", "agent"]
first,second,*rest = list
print(first)
print(second)
print(*rest)

#slicing items from a list
fruits = ['banana', 'orange', 'mango', 'lemon', 'grape']
print(fruits[0:4])
print(fruits[::2])
#
game = ['canva','unity', 'twitter', 'engine', 'oil', 'lame', 'start']
print(game[::2])
print(game[::3])
print(game[:2])
print(game[1:3])

#negative indexing
print(game[-3:-1])
print(game[-3:])
print(game[::-1])

#modifying in list
fruits = ['banana', 'orange', 'mango', 'lemon']
fruits[0] = "avogado"
print(fruits)

#
name = ['seun', 'kemi', 'kuti', 'segun']
name[3] = 'eyitayo'
print(name)
name[0] ='fiyin'
print(name)
```

```python
#checking items in a list
fruits = ['banana', 'orange', 'mango', 'lemon']
name = 'banana' in fruits
print(name)

#adding to items in list To add item to the end of an existing list we use the method append().
name = ['kemi', 'kemo', 'kamsi', 'kay']
name.append('funmi')
print(name)

# Insert; it helps to insert and item into a specified index
name = ['kim', 'john', 'wick', 'frontier']
name.insert(2, 'yemi')
print(name)

#remove format
name.remove('kim')
print(name)

#pop format
name.pop(2)
print(name)

#Joining in list
#plus operator
list1 = [1,2,3,4,5]
list2 = ['femi', 'yemi']
print(list1+list2)

#extend method
list1.extend(list2)
print(list1)

#count
name = [1,2,3,4,5,4,4,4,4]
print(name.count(4))
print(name.index(3))
#reverse
guy = [1, 2, 3, 4, 5]
guy.reverse()
print(guy)

#sort
letter = ['c', 'a', 'd', 'b']
letter.sort()
print(letter)
```
