```python
##Level 2

# 6a
file = open(r"C:\\Users\\User\\Downloads\\Obasi Python\\obama_speech.txt")   
text = file.read()  
#print(text)
def top_10_mfw(file_path):
    with open(file_path, 'r') as file:
        words = file.read().split()
        word_freq = {}
        for word in words:
            word_freq[word] = word_freq.get(word, 0) + 1
        sorted_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)
        return sorted_words[:3]

top_words = top_10_mfw("C:\\Users\\User\\Downloads\\Obasi Python\\obama_speech.txt")
if top_words:
    print("Top 3 most frequent words:")
    for word, freq in top_words:
        print(f"{word}: {freq}")

file.close()
# 6b
file = open(r"C:\\Users\\User\\Downloads\\Obasi Python\\michelle_obama_speech.txt", encoding='utf-8')   

# text = file.read()  
#print(text)
def top_10_mfw(file_path):
    with open(file_path, 'r') as file:
        words = file.read().split()
        word_freq = {}
        for word in words:
            word_freq[word] = word_freq.get(word, 0) + 1
        sorted_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)
    return sorted_words[:3]

top_words = top_10_mfw("C:\\Users\\User\\Downloads\\Obasi Python\\michelle_obama_speech.txt", encoding='utf-8')
if top_words:
    print("Top 3 most frequent words:")
    for word, freq in top_words:
        print(f"{word}: {freq}")

file.close()
```
