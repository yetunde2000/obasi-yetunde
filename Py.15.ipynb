```python
# Exercises: Day 20
#1
url = 'http://www.gutenberg.org/files/1112/1112.txt'
response = requests.get(url)
def top_10_mfw(response):
    words = response.text.split()
    word_freq = {}
    for word in words:
        word_freq[word] = word_freq.get(word, 0) + 1
    sorted_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)
    return sorted_words[:3]

top_words = top_10_mfw(response)
if top_words:
    print("Top 3 most frequent words:")
    for word, freq in top_words:
        print(f"{word}: {freq}")
        
# 2 i
from statistics import mean

url = 'https://api.thecatapi.com/v1/breeds'
response = requests.get(url)
data = response.json()

# Assuming data is a list of dictionaries and we want to find min, max, and mean of a specific numeric field
# For example, let's assume we are interested in the 'weight' field of each breed
weights = [breed['weight']['metric'] 
           for breed in data
           if 'weight' in breed and 'metric' in breed['weight']]

# Extract the average weight for each breed
weights = [(float(w.split(' - ')[0]) + float(w.split(' - ')[1])) / 2 for w in weights]

print(min(weights))
print(max(weights))
print(mean(weights))

#ii
import requests
from statistics import mean

url = 'https://api.thecatapi.com/v1/breeds'
response = requests.get(url)
data = response.json()
lifespans = [breed['life_span'] 
           for breed in data
           if 'life_span' in breed]

# Extract the average lifespan for each breed
lifespans = [(float(w.split(' - ')[0]) + float(w.split(' - ')[1])) / 2 for w in lifespans]

print(min(lifespans))
print(max(lifespans))
print(mean(lifespans))

#iii
import requests
import pandas as pd

# Install html5lib
%pip install html5lib

url = 'https://api.thecatapi.com/v1/breeds'
response = requests.get(url)
data = response.json()

# Create a frequency table of country and breed of cats
country_breed_freq = {}
for breed in data:
    country = breed.get('origin', 'Unknown')
    country_breed_freq[country] = country_breed_freq.get(country, 0) + 1

# Convert the frequency table to a DataFrame
df = pd.DataFrame(list(country_breed_freq.items()), columns=['Country', 'Breed Count'])
print(df)
```
