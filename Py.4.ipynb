```python
#30dayspythonchallenge_Day8
#Dictionaries
#A dictionary is a collection of unordered, modifiable(mutable) paired (key: value) data type.
# we use ths "{}" in creating a dictionary

#
person = {'Name':'Yetunde',
          'Age': 24,
          'School': 'DGIST',
          'is_married': True,
          'Skills': ['CAD designer', 'programmer']}
print(person)
print(len(person))

#Accessing item in dictionary
print(person['Name'])
print(person['School'])
print(person['Skills'][0])

# Adding items to a Dictionary.
person['Friends'] = ['KEMI','Smith']
print(person)

#modify
person['Name'] = 'Micheal'
print(person)

#checking keys in dictionary
print('Name' in person)
print('Animal' in person)

#remove
person.pop('Name')
print(person)
del person['Age']
print(person)

#changing dict to List items
print(person.items())
#to clear
print(person.clear())

#getting dict keys as list
keys = person.keys()
print(keys)
# dicts values as list
values = person.values()
print(values)
```
