```python
#30daysPythonchallenge_DAY10
#LOOPS
#loop is a like a life routine full of repetitive task and to handle this task, we use the loop.
#while loop and For loop
#while condition continues to execute untill the comdition is satisfied
#while condition
      #code
#
count =  0
while count > 5:
    print(count)
    count += 1
else:
    print(count)
#
sum = 5
while sum < 15:
    print(sum)
    sum += 1
else:
    print(sum)
```

```python
#break and continue
#we break wen we like to stop a loop
#while condition
     #code
     #if condition
        #break
count = 0
while count < 5:
    print(count)
    count += 1
    if count == 3:
        break
    
#continue
count = 0
while count < 5:
    if count == 2:
        count += 1
        continue
    print(count)
    count += 1 #it skips 3
#
count = 10
while count > 15:
    if count == 12:
        count += 2
        continue
    print(count)
    count += 2
    
```

```python
#For Loop
#for loop
    #code

#
num = [0, 1, 2, 4, 5]
for i in num:
    print(i)

#
name = ['kemi', 'segun','tunde', 'funmi','funto']
for names in name:
    print(names)
#3 in strings
nam = 'Python'
for i in range(len(name)):
    print(nam[i])
# for; for tuple
num = (0, 1, 2, 3, 4, 5, 6)
for i in num:
    print(i)
    
# dict
num = {'key':'kim',
       'keey': ' kim2'}
for key in num:
    print(key)
    
#set
comp = {'Facebook', 'Google', 'Microsoft', 'Apple', 'IBM', 'Oracle', 'Amazon'}
for i in comp:
    print(i)
```

```python
#Break and continue
num = (0, 1, 2, 3, 4, 5)
for i in num:
    print(i)
    if i == 2:
        break
    
#
num = (0,1,2,3,4,5)
for i in num:
    print(i)
    if i == 3:
        continue
    if i != 5:
        print(i + 1)
    else:
        print('okay')
print('outside the loop')
```
