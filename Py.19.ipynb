```python
# Day 25 - Pandas
# andas is an open source, high-performance, easy-to-use data structures and data analysis tools for the Python programming language. 
#reshaping
#merging
#sorting
#slicing
#aggregation
#imputation. If you are using anaconda, you do not have install pandas.


#Installing Pandas
#pip install conda # if you are using anaconda, you do not have to install pandas
#pip install pandas
    #Names Pandas Series
    #import pandas as pd
import pandas as pd
nums = [1,2,3,4,5]
s = pd.Series(nums)
print(s)

numss = [1,2,3,4,5]
s = pd.Series(numss, index=['a','b','c','d','e'])
print(s)

fruits = ['apple', 'banana', 'cherry']
s = pd.Series(fruits, index=['a','b','c'])
print(s)

# creating pandas from a dictionary
dict = {"a": "Kemi", "country": "Nigeria", "age": 25}
print(pd.Series(dict))

# Creating Dataframe from list of lists
data = [["Kemi", "Oguntolu", "Lagos"],
        ["seun", "tanke", "ilorin"],
        ["Eni", "ayedaide", "ikire"],
        ["Tolu", "pele", "ondo"]]
df = pd.DataFrame(data, columns=["Name", "city", "State"])
print(df)

# creating Dataframe from dictionary
data = [{"Name": "Kemi", "city": "Lagos", "State": "Nigeria"},
        {"Name": "seun", "city": "ilorin", "State": "Nigeria"},
        {"Name": "Eni", "city": "ikire", "State": "Nigeria"},
        {"Name": "Tolu", "city": "ondo", "State": "Nigeria"}]
df = pd.DataFrame(data)

print(df)
```
