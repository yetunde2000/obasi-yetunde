```python
#
from scipy import stats
from scipy.stats import mode
import matplotlib.pyplot as plt
nppp = np.random.normal(5, 0.2, 50)
print(np)
print(np.min(nppp))
print(np.max(nppp))
print(np.std(nppp))
print(np.var(nppp)) 
print(np.mean(nppp))
print(np.median(nppp))
#print(np.stats(mode(nppp)))
plt.hist(nppp, color = "red", bins=21)
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

```

```python
# Linear algebra in numpy
# 1. Dot product
f = np.array([1,2,3])
g = np.array([4,5,6])
print(np.dot(f,g))

# 2. Matrix multiplication
f = np.array([[1,2],[3,4]])
g = np.array([[5,6],[7,8]])
print(np.matmul(f,g))
print(np.dot(f,g))

## linear equation
# 1. Solve linear equations
temp =np.array([1,2,3,4])
pressure = temp * 2 + 3
print(pressure)
plt.plot(temp, pressure)
plt.title("Linear Equation")
plt.xlabel("Temperature")
plt.ylabel("Pressure")
plt.show()

# gaussian distrubution
import seaborn as sns
mu = 25
sigma = 10
samples = 1000
x = np.random.normal(mu, sigma, samples)
ax = sns.distplot(x);
ax.set(xlabel='Value', ylabel='Frequency')
plt.title("Gaussian Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

#Arrays support vectorized operations, while lists dont.
#Once an array is created, you cannot change its size. You will have to create a new array or overwrite the existing one.
#Every array has one and only one dtype. All items in it should be of that dtype.
#An equivalent numpy array occupies much less space than a python list of lists.
#numpy arrays support boolean indexing.
```
